```python
df['boolquality'] = np.where(df.quality>5, 1, 0)
```

```python
tf_data = torch.tensor(df[cols].values).float()
tf_labels = torch.tensor(df.boolquality).float()
print(tf_data.shape)
print(tf_labels.shape)
```

```python
# We need to convert the labels to one dimensional tensor
tf_labels = tf_labels[:, None]
print(tf_labels.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(tf_data, tf_labels, train_size=0.7)
train_data = TensorDataset(x_train, y_train)
test_data = TensorDataset(x_test, y_test)

batch = 64
train_loader = DataLoader(train_data, batch_size=batch, shuffle=True, drop_last=True)
test_loader = DataLoader(test_data, batch_size=test_data.tensors[0].shape[0])
```

# Model

```python
def model():
    ann = torch.nn.Sequential(
        torch.nn.Linear(11, 64),
        torch.nn.ReLU(),
        torch.nn.Linear(64,64),
        torch.nn.ReLU(),
        torch.nn.Linear(64, 1)
    )
    lossfn = torch.nn.BCEWithLogitsLoss()
    optimizer = torch.optim.SGD(ann.parameters(), lr=0.01)
    return ann, lossfn, optimizer
```

```python
ann, lossfn, optimizer = model() 
```

```python
ann
```

```python
epochs = 100
def train():
    train_acc = []
    test_acc = []
    losses = []
    for i in range(epochs):
        ann.train()
        batchacc = []
        batchloss = []
        for xtrain, ytrain in train_loader:
            yhat = ann(xtrain)
            loss = lossfn(yhat, ytrain)
            batchloss.append(loss.detach().numpy())
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            batchacc.append(100*torch.mean(((yhat>0) == ytrain).float()).item())
        train_acc.append(np.mean(batchacc))
        losses.append(np.mean(batchloss))
        ann.eval()
        xtest, ytest = next(iter(test_loader))
        with torch.no_grad():
            pred = ann(xtest)
        test_acc.append(100*torch.mean(((pred>0) == ytest).float()).item())

    return train_acc, test_acc, losses
```

```python
train_acc, test_acc, losses = train()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12,4))
ax[0].plot(range(epochs), train_acc)
ax[0].plot(range(epochs), test_acc)
ax[0].set_xlabel('epochs')
ax[0].set_ylabel('accuracy')
ax[0].legend(['train_acc', 'test_acc'])

ax[1].plot(range(epochs), losses)
ax[1].set_xlabel('epochs')
ax[1].set_ylabel('losses')
```

# Experiment

```python
bsizes = 2**np.arange(1, 10, 2)
```

```python
bsizes
```

```python
exp_train_acc = np.zeros((epochs, len(bsizes)))
exp_test_acc = np.zeros((epochs, len(bsizes)))
exp_losses = np.zeros((epochs, len(bsizes)))
```

```python
for i in range(len(bsizes)):
    train_loader = train_loader = DataLoader(train_data, batch_size=int(bsizes[i]), shuffle=True, drop_last=True)
    ann, lossfn, optimizer = model() 
    train_acc, test_acc, losses = train()
    exp_train_acc[:,i] = train_acc
    exp_test_acc[:,i] = test_acc
    exp_losses[:,i] = losses
```

```python
fig, ax = plt.subplots(2, 2, figsize=(14,6))
ax[0,0].plot(range(epochs), exp_train_acc)
ax[0,0].set_xlabel('epochs')
ax[0,0].set_ylabel('accuracy')
ax[0,0].legend(bsizes)

ax[0,1].plot(range(epochs), exp_test_acc)
ax[0,1].set_xlabel('epochs')
ax[0,1].set_ylabel('accuracy')
ax[0,1].legend(bsizes)

ax[1,0].plot(range(epochs), exp_losses)
ax[1,0].set_xlabel('epochs')
ax[1,0].set_ylabel('losses')
ax[1,0].legend(bsizes)
```
