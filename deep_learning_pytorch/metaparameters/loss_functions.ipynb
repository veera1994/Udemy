```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
```

# MSE

```python
lossfn = nn.MSELoss()
yhat = torch.linspace(-2, 2, 101)
y = torch.tensor(0.5)
l = np.zeros(101)
```

```python
for i in range(len(yhat)):
    l[i] = lossfn(yhat[i], y)
```

```python
plt.plot(yhat, l)
plt.xlabel('predicted value')
plt.legend()
```

# BCE

```python
lossfn = nn.BCELoss()
yhat = torch.linspace(.0001, .999, 101)
y1 = torch.tensor(0.)
y2 = torch.tensor(1.)
l = np.zeros((101, 2))
```

```python
for i in range(len(yhat)):
    l[i, 0] = lossfn(yhat[i], y1)
    l[i, 1] = lossfn(yhat[i], y2)
```

```python
plt.plot(yhat, l)
plt.xlabel('predicted value')
plt.legend(['0', '1'])
```

```python
yhat = torch.tensor(3.0)
print(lossfn(nn.Sigmoid()(yhat), y2))
# When you use BCEloss you need to do the sigmoid transformation
```

```python
lossfn = nn.BCEWithLogitsLoss()
yhat = torch.tensor(3.0)
print(lossfn(yhat, y2))
# BCE logit loss is BCE with sigmoid function
```

# CCE

```python
lossfn = nn.CrossEntropyLoss()
yhat = torch.tensor([0.1, 14.00001, 4.])
for i in range(len(yhat)):
    correct_answer = torch.tensor(i)
    loss = lossfn(yhat, correct_answer)
    print(f"loss: {loss} when correct answer is {correct_answer}")
```

```python
yhat = torch.tensor([[0.1, 3, 4.]])
print(yhat)
print(nn.Softmax(dim=1)(yhat))
print(nn.LogSoftmax(dim=1)(yhat))
```

```python
# Log softmax is better than softmax because it stretches the values near the zero
```
